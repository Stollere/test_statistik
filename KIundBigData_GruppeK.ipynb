Each row in the dataset describes a country in a given year. Independent variables include various attributes on health, education, economic status and population. The dependent variable of this dataset is the life expectancy in years. Our goal will be to build a model to predict the life expectancy of any given country.

If we take a look at the types we can see that we are mostly dealing with numerical attributes so we won't have to do much preprocessing in terms of encoding.

```python
life_expectancy.info()
```

Actually, the only exception is the status attribute. To improve handling, we will use a simple label encoder:
(Transformation der nominalen Variable in eine binäre Variable) --> Dummys

Country ist ein Problem, da nicht numerisch. --> brauchen wir nicht für Vorhersage, daher wird der nicht umkodiert. 

```python
life_expectancy.Status.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
life_expectancy.Status = encoder.fit_transform(life_expectancy.Status)
```

Transformation vom Typ Object zum Integer

```python
life_expectancy.Status.value_counts()
```

Interesting statistics of our attributes:

```python
life_expectancy
```

By taking a closer look at null values for each attribute we can see that there is quite a big difference in completeness:

```python
life_expectancy.isnull().sum()
```

While this information already is quite useful, we always have to put the number of missing values into relation to the absolute number of records in the dataset.

For this, we can just divide the numbers from above by the overall number of rows:



```python
life_expectancy.isnull().sum()/len(life_expectancy)
```

We can also format these numbers as percentages to make them easier to interpret:

```python
(life_expectancy.isnull().sum()/len(life_expectancy)).apply('{0:.3%}'.format)
```

We'll be using some algorithms that cannot deal with null values, so we will have to remove them somehow. We could use a more sophisticated approach like filling null values with the mean or median of the attributes but our dataset has a sufficient size even after dropping the rows with missing values, so it seems feasible to simply drop these rows.

```python
life_expectancy = life_expectancy.dropna()
```

```python
life_expectancy.isnull().sum()
```

Überprüfung passt. Es sind keine Nullwerte mehr vorhanden. 

```python
len(life_expectancy)
```
